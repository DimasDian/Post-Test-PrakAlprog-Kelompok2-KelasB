KATEGORI A

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Kelas B_CO2 Emissions by Car.csv", sep=';')

max_engine = df.groupby('MAKE')['ENGINE_SIZE'].max()

top_10_engine = max_engine.sort_values(ascending=False).head(10)

print("10 Merek Teratas Berdasarkan Engine Size Maksimum:")
print(top_10_engine)

fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(
    x=top_10_engine.index,
    y=top_10_engine.values,
    ax=ax
)

ax.set_title('Kategori A: 10 Merek Mobil dengan Engine Size Maksimum Terbesar')
ax.set_xlabel('Merek Mobil')
ax.set_ylabel('Engine Size Maksimum')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

KATEGORI B

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Kelas B_CO2 Emissions by Car.csv", sep=';')

filtered_df = df[
    (df['TRANSMISSION'].isin(['A5', 'AS5'])) &
    (df['CYLINDERS'] > 4)
]

print("Data mobil transmisi A5/AS5 dengan silinder > 4:")
display(filtered_df)

avg_emission = filtered_df.groupby('MAKE')['CO2_EMISSIONS'].mean()

avg_emission = avg_emission.sort_values(ascending=False)

print("Rata-rata Emisi CO2 per Merek:")
print(avg_emission)

fig, ax = plt.subplots(figsize=(14, 6))

sns.barplot(
    x=avg_emission.index,
    y=avg_emission.values,
    ax=ax
)

ax.set_title('Kategori B: Rata-rata Emisi CO2 Mobil Transmisi A5/AS5 dan Silinder > 4')
ax.set_xlabel('Merek Mobil')
ax.set_ylabel('Rata-rata Emisi CO2')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

KATEGORI C

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Kelas B_CO2 Emissions by Car.csv", sep=';')

fig, ax = plt.subplots(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x='CYLINDERS',
    y='FUEL_CONSUMPTION*',
    ax=ax
)

ax.set_title('Kategori C: Hubungan Jumlah Silinder dan Konsumsi Bahan Bakar')
ax.set_xlabel('Jumlah Silinder')
ax.set_ylabel('Konsumsi Bahan Bakar')

plt.tight_layout()
plt.show()
```

KATEGORI D

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Kelas B_CO2 Emissions by Car.csv", sep=';')

volvo_df = df[df['MAKE'] == 'VOLVO']

print("Data mobil merek VOLVO:")
display(volvo_df)

fig, ax = plt.subplots(figsize=(8, 6))

sns.histplot(
    data=volvo_df,
    x='FUEL_CONSUMPTION*',
    kde=True,
    ax=ax
)

ax.set_title('Kategori D: Sebaran Konsumsi Bahan Bakar Mobil VOLVO')
ax.set_xlabel('Konsumsi Bahan Bakar')
ax.set_ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

GRAFIK GABUNGAN

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Kelas B_CO2 Emissions by Car.csv", sep=';')

fig, axes = plt.subplots(2, 2, figsize=(20, 14))

max_engine = df.groupby('MAKE')['ENGINE_SIZE'].max()
top_10_engine = max_engine.sort_values(ascending=False).head(10)

sns.barplot(
    x=top_10_engine.index,
    y=top_10_engine.values,
    ax=axes[0, 0]
)

axes[0, 0].set_title('Kategori A: Top 10 Engine Size Maksimum')
axes[0, 0].set_xlabel('Merek Mobil')
axes[0, 0].set_ylabel('Engine Size Maksimum')
axes[0, 0].tick_params(axis='x', rotation=45)

filtered_df = df[
    (df['TRANSMISSION'].isin(['A5', 'AS5'])) &
    (df['CYLINDERS'] > 4)
]

avg_emission = filtered_df.groupby('MAKE')['CO2_EMISSIONS'].mean()
avg_emission = avg_emission.sort_values(ascending=False)

sns.barplot(
    x=avg_emission.index,
    y=avg_emission.values,
    ax=axes[0, 1]
)

axes[0, 1].set_title('Kategori B: Rata-rata Emisi CO2 A5/AS5')
axes[0, 1].set_xlabel('Merek Mobil')
axes[0, 1].set_ylabel('Rata-rata Emisi CO2')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.scatterplot(
    data=df,
    x='CYLINDERS',
    y='FUEL_CONSUMPTION*',
    ax=axes[1, 0]
)

axes[1, 0].set_title('Kategori C: Silinder vs Konsumsi Bahan Bakar')
axes[1, 0].set_xlabel('Jumlah Silinder')
axes[1, 0].set_ylabel('Konsumsi Bahan Bakar')

volvo_df = df[df['MAKE'] == 'VOLVO']

sns.histplot(
    data=volvo_df,
    x='FUEL_CONSUMPTION*',
    kde=True,
    ax=axes[1, 1]
)

axes[1, 1].set_title('Kategori D: Distribusi Konsumsi BBM VOLVO')
axes[1, 1].set_xlabel('Konsumsi Bahan Bakar')
axes[1, 1].set_ylabel('Frekuensi')

fig.suptitle('Grafik Gabungan Pengolahan Data CO2 Emissions by Car', fontsize=18)

plt.tight_layout()
plt.show()
```
